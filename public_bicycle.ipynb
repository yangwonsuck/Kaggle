```python
import datetime # 날짜를 추출하기 위한 라이브러리 선언
df['year']=pd.DatetimeIndex(df['datetime']).year  #연도를 따로 뽑아냄 
df['month']=pd.DatetimeIndex(df['datetime']).month #월을 따로 뽑아냄
df['day']=pd.DatetimeIndex(df['datetime']).day #날짜를 따로 뽑아냄
df['hour']=pd.DatetimeIndex(df['datetime']).hour #시간을 따로 뽑아냄
df['minute']=pd.DatetimeIndex(df['datetime']).minute #분을 따로 뽑아냄
df['second']=pd.DatetimeIndex(df['datetime']).second #초를 따로 뽑아냄
df['dayofweek']=pd.DatetimeIndex(df['datetime']).dayofweek #요일을 따로 뽑아냄 0부터 6까지 (월~일순)
```

# 조별과제

# 1.시간별(년,월,일,시간,분,초)에 따른 대여량을 시각화(x축 시간값,y축 대여량)
# 6가지 시각화된 데이터에 대한 특징적 요소 서술

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('dark')

fig=plt.figure(figsize=(15,10))
ax1=fig.add_subplot(2,3,1)
ax2=fig.add_subplot(2,3,2)
ax3=fig.add_subplot(2,3,3)
ax4=fig.add_subplot(2,3,4)
ax5=fig.add_subplot(2,3,5)
ax6=fig.add_subplot(2,3,6)

sns.barplot(x='year',y='count',data=df,ax=ax1) #2012년이 2011년 보다 대여수가 더 많다.
sns.barplot(x='month',y='count',data=df,ax=ax2) #6~9월 까지 즉 여름기간에 대여수가 더 많다.
sns.barplot(x='day',y='count',data=df,ax=ax3)  
sns.barplot(x='hour',y='count',data=df,ax=ax4) #오후 5~6시에 대여수가 급등한다.
sns.barplot(x='minute',y='count',data=df,ax=ax5) 
sns.barplot(x='second',y='count',data=df,ax=ax6)

ax1.set_title('bicycle year-count')
ax2.set_title('bicycle month-count')
ax3.set_title('bicycle day-count')
ax4.set_title('bicycle hour-count')
ax5.set_title('bicycle minute-count')
ax6.set_title('bicycle second-count')

plt.show()
```

# 2.전체 대여량,계절별 대여량,시간별 대여량,근무일에 따른 대여량 시각화

```python
sns.set_style('dark')

fig=plt.figure(figsize=(15,10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)


sns.barplot(x='day',y='count',data=df,ax=ax1) 
sns.barplot(x='season',y='count',data=df,ax=ax2) #여름,가을에 대여수가 많은 것을 확인할수 있다.
sns.barplot(x='hour',y='count',data=df,ax=ax3)   #오후 5~6시에 대여수가 급등한다.
sns.barplot(x='workingday',y='count',data=df,ax=ax4) #평일에 살짝 대여수가 더 많음(상대적인 일수 비교하면 휴일이 훨씬 많음)


ax1.set_title('bicycle day-count')
ax2.set_title('bicycle season-count')
ax3.set_title('bicycle hour-count')
ax4.set_title('bicycle workingday-count')

plt.show()
```

# 7.온도와 등록 사용자 여부,습도,풍속 데이터가 어떤 연관관계가 있는지 히트맵 시각화

```python
# temp,registered,humidity,windspeed
sns.set_style('dark')
fig=plt.figure(figsize=(15,10))

table = df.pivot_table(index=['registered'],columns=['windspeed'],aggfunc='size')

sns.heatmap(table,
           annot=True,fmt='f', #데이터 값 표시 여부, 실수형 포맷
           cmap='YlGnBu',  #컬러 맵
           linewidth=.5,  #구분선
           cbar=True) #컬러바 표시여부

plt.show() #풍속이 낮을수록 등록된 사용자가 많음
```

```python
# temp,registered,humidity,windspeed

fig=plt.figure(figsize=(15,10))

table = df.pivot_table(index=['registered'],columns=['humidity'],aggfunc='size')

sns.heatmap(table,
           annot=True,fmt='f',
           cmap='YlGnBu',
           linewidth=.5,
           cbar=False)

plt.show() #습도가 50이상일 때 등록된 사용자가 많음
```

```python
# temp,registered,humidity,windspeed

fig=plt.figure(figsize=(15,10))

table = df.pivot_table(index=['registered'],columns=['temp'],aggfunc='size')

sns.heatmap(table,
           annot=True,fmt='f',
           cmap='YlGnBu',
           linewidth=.5,
           cbar=False)

plt.show() #날씨가 따듯할수록 등록된 사용자가 많음
```

# 8.온도,풍속,습도에 대한 산점도를 시각화로 표현(regplot)

```python
sns.set_style('ticks')

fig=plt.figure(figsize=(15,10))
ax1=fig.add_subplot(2,4,1)
ax2=fig.add_subplot(2,4,2)
ax3=fig.add_subplot(2,4,3)

sns.regplot(x='temp',y='count',data=df,ax=ax1,color='r')  #온도가 20~30일 때 대여수가 많음
sns.regplot(x='windspeed',y='count',data=df,ax=ax2,color='g') #풍속이 10~20일 때 대여수가 많음 
sns.regplot(x='humidity',y='count',data=df,ax=ax3,color='b')  #습도가 40~60일 때 대여수가 많음
plt.show()
```
